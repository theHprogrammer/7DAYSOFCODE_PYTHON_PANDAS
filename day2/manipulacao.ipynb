# Dia 1 - Importação de Dados

```python
# Bibliotecas
import pandas as pd
import os
```

```python
# Pegar os caminhos dentro da pasta data
def get_data_paths(folder_path: str):
    data_folder = os.path.join(os.getcwd(), folder_path)
    data_paths = [os.path.join(data_folder, file) for file in os.listdir(data_folder)]
    return data_paths

# Ler os arquivos csv
def read_csv_files(data_paths: list):
    data_frames = [pd.read_csv(file) for file in data_paths]
    return data_frames
```

#### Lendo os arquivos de emprestimos e unindo-os em um único dataframe

```python
df_emprestimos = pd.concat(read_csv_files(get_data_paths('../data/csv/')), ignore_index=True)
df_emprestimos
```

#### Verifica se há valores duplicados

```python
df_emprestimos.duplicated().sum()
```

#### Exclui as duplicatas

```python
df_emprestimos = df_emprestimos.drop_duplicates()
```

```python
df_emprestimos.duplicated().sum()
```

#### Verificando valores nulos

```python
df_emprestimos.isnull().sum()
```

#### Carregando os exemplares

```python
df_exemplares = pd.read_parquet('../data/dados_exemplares.parquet')
df_exemplares
```

#### Verificando duplicatas para os exemplares

```python
df_exemplares.duplicated().sum()
```

#### Mesclando emprestimos com exemplares

```python
df = pd.merge(df_emprestimos, df_exemplares, on='codigo_barras', how='left')
df
```

#### Aplicar a coluna CDU

- 000 a 099: Generalidades. Ciência e conhecimento.
- 100 a 199: Filosofia e psicologia.
- 200 a 299: Religião.
- 300 a 399: Ciências sociais.
- 400 a 499: Classe vaga. Provisoriamente não ocupada.
- 500 a 599: Matemática e ciências naturais.
- 600 a 699: Ciências aplicadas.
- 700 a 799: Belas artes.
- 800 a 899: Linguagem. Língua. Linguística.
- 900 a 999: Geografia. Biografia. História."

```python
cdu_list = []
for cdu in df['localizacao']:
    if cdu < 100:
        cdu_list.append('Generalidades. Ciência e conhecimento.')
    elif cdu < 200:
        cdu_list.append('Filosofia e psicologia.')
    elif cdu < 300:
        cdu_list.append('Religião.')
    elif cdu < 400:
        cdu_list.append('Ciências sociais.')
    elif cdu < 500:
        cdu_list.append('Classe vaga. Provisoriamente não ocupada.')
    elif cdu < 600:
        cdu_list.append('Matemática e ciências naturais.')
    elif cdu < 700:
        cdu_list.append('Ciências aplicadas.')
    elif cdu < 800:
        cdu_list.append('Belas artes.') 
    elif cdu < 900:
        cdu_list.append('Linguagem. Língua. Linguística.')
    else:
        cdu_list.append('Geografia. Biografia. História.')
        
df['cdu_geral'] = cdu_list
df.head(5)  

```
